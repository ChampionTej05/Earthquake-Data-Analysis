## Plotting the earthquakes on the map

```python
import folium
import pandas as pd


map = folium.Map(location=[51.578939, 58.022448], zoom_start=2, tiles="Mapbox Bright")

fgd=folium.FeatureGroup(name='Above 8')

data1=pd.read_csv('project/earthquake_japan_8.csv')
lat=list(data1['Latitude'])
lon=list(data1['Longitude'])
depth=list(data1['Depth'])
elev=list(data1['Magnitude'])

def color_producer(elevation):
    if elevation >=9.00:
        return 'red'
    elif 8.5 <= elevation < 9.00:
        return 'blue'
    else:
        return 'white'

def radius_producer(dep):
    if dep > 100:
        return 2
    elif 50 <= dep <100:
        return 4
    else:
        return 6
    
for lt, ln, el,dep in zip(lat, lon, elev,depth):
    fgd.add_child(folium.CircleMarker(location=[lt, ln], radius = radius_producer(dep), popup=str(dep),
    fill_color=color_producer(el), fill=True,  color = 'grey', fill_opacity=0.7))
    
    

map.add_child(fgd)

map.add_child(folium.LayerControl())
map.save('project/map_earthquake_japan_8.html')
```

## Creating the dataset

```python
df1=pd.DataFrame(columns=['Latitude','Longitude','Date','Depth','Magnitude','Trig_earthquakes'])
```

```python
def isRegion1(lat,lon):
    if region1[0][0]>= lat >= region1[2][0]:
        if region1[0][1]<=  lon <= region1[2][1]:
            return True 
    return False

def isRegion2(lat,lon):
    if region2[0][0]>= lat >= region2[2][0]:
        if region2[0][1]<=  lon <= region2[2][1]:
            return True 
    return False

def isRegion3(lat,lon):
    if region3[0][0]>= lat >= region3[2][0]:
        if region3[0][1]<=  lon <= region3[2][1]:
            return True 
    return False

def isRegion4(lat,lon):
    if region4[0][0]>= lat >= region4[2][0]:
        if region4[0][1]<=  lon <= region4[2][1]:
            return True 
    return False

def check(choice,lat,lon):
    if choice==1:
        isRegion1(lat,lon)
    elif choice==2:
        isRegion2(lat,lon)
    elif choice==3:
        isRegion3(lat,lon)
    elif choice==4:
        isRegion4(lat,lon)

def return_choice(lat,lon):
    if isRegion1(lat,lon):
        return 1
    elif isRegion2(lat,lon):
        return 2
    elif isRegion3(lat,lon):
        return 3
    elif isRegion4(lat,lon):
        return 4
    
def inNextYear(date1,date2):
    from datetime import datetime as dt
    d1=dt.strptime(date1,'%m/%d/%Y')
    d2=dt.strptime(date2,'%m/%d/%Y')
    d_date=d2.year-d1.year
    if d_date ==1 or d_date==0:
        return True
    else: 
        return False
```

```python
row_list=list()
index_df=0
for index,row in df.iterrows():
    cur_lat,cur_lon=row['Latitude'],row['Longitude']
    print('Current Earthquake : ',(row['Latitude'],row['Longitude'],row['Depth'],row['Date']))
    row_list=list()
    choice=return_choice(cur_lat,cur_lon)
    print("current choice: ",choice)
    print('----------------------------------')
    count =0
    for indext,rowt in df.iterrows():
        
        temp_lat,temp_lon=rowt['Latitude'],rowt['Longitude']
        if (temp_lat != cur_lat) and (cur_lon != temp_lon):
            if not check(choice,temp_lat,temp_lon):
                if inNextYear(row['Date'],rowt['Date']):
                    count =count +1
                    print("- - - - - - - - - - - - - - - ")
                    print('Earthquake: ',temp_lat,temp_lon)
                    print('Trigerred by: ',cur_lat,cur_lon)
                    print('Year of Earthquake: ',rowt['Date'])
                    print("- - - - - - - - - - - - - - - ")
    print('Total Earthquakes Triggered: ',count)
    d={'Latitude':cur_lat,'Longitude':cur_lon,'Date':row['Date'],'Depth':row['Depth'],'Magnitude':row['Magnitude'],'Trig_earthquakes':count}
    df1.loc[index_df,:]=d
    index_df=index_df+1
    
    print('----------------------------------------')
    
```
