## Plotting all the earthquakes

````python
#filetring it 
import folium
import pandas as pd

map = folium.Map(location=[38.58, -99.09], zoom_start=6, tiles="Mapbox Bright")
fgd=folium.FeatureGroup(name='Earthquakes')

data1=df
lat=list(data1['Latitude'])
lon=list(data1['Longitude'])
depth=list(data1['Depth'])
elev=list(data1['Magnitude'])

def color_producer(elevation):
    if elevation >6.00:
        return 'red'
    elif 5.8 <= elevation < 6.00:
        return 'blue'
    else:
        return 'white'

def radius_producer(dep):
    if dep > 60:
        return 8
    elif 50 <= dep <60:
        return 3
    else:
        return 2
    
for lt, ln, el,dep in zip(lat, lon, elev,depth):
    fgd.add_child(folium.CircleMarker(location=[lt, ln], radius = radius_producer(dep), popup=str(el),
    fill_color=color_producer(el), fill=True,  color = 'grey', fill_opacity=0.7))
    
map.add_child(fgd)
map.add_child(folium.LayerControl())
map.save('project/map_earthquakes_world.html')
````

### Between 2005 to 2009

````python
df_5_to_9=pd.DataFrame(columns=['Latitude','Longitude','Date','Depth','Magnitude'])
````

````python
from datetime import datetime as dt
````

````python

row_list=list()
count=0
for index,row in df.iterrows():
    try:
        if dt.strptime('01/01/2005',"%m/%d/%Y")<=dt.strptime(row['Date'],"%m/%d/%Y")<=dt.strptime('12/31/2009',"%m/%d/%Y"):
            d={'Latitude':row['Latitude'],'Longitude':row['Longitude'],'Date':row['Date'],'Depth':row['Depth'],'Magnitude':row['Magnitude']}
            df_5_to_9.loc[count,:]=d
            count=count+1
            
    except Exception as e:
        print(e)
````

````python
df_5_to_9.to_csv('project/earthquakes_2005_2009.csv')
````
